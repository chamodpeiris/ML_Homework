```python
# IMPORT YOUR LIBRARY HERE

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# IMPORT THE DATASET FROM GITHUB LINK

url = "https://raw.githubusercontent.com/afriwahyudi/datasets/main/extractive_distillation/5solvents_ExDis.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
# PICKING A SOLVENT AND FILTERING STRINGS FROM THE DATASET

# Identify all the solvent names
solvent_names = df['Solvent'].unique()
print("All Solvent Names:", solvent_names)

# Pick a solvent name
chosen_solvent = '2-4PDN'
filtered_df = df[df['Solvent'] == chosen_solvent]
filtered_df.head(10)
```

```python
# GENERATE A HEATMAP OF THE NUMERICAL DATASETS (NO NEED TO SEPARATE FEATURES AND TARGETS)

# Filter numerical columns
numerical_columns = filtered_df.select_dtypes(include=['float64', 'int64']).columns

# Generate Heatmap
plt.figure(figsize=(13, 6))
heatmap_data = filtered_df[numerical_columns].corr()

cmap = sns.diverging_palette(120,320, s=100, l=20, n=10, as_cmap=True)
sns.heatmap(heatmap_data, annot=True, cmap=cmap, fmt=".2f", linewidths=2)
plt.title(f"Heatmap for {chosen_solvent} Features")
plt.show()
```

Lastly, interpret the heatmap!

**The correlation heatmap for 2-4PDN is shown in Figure.**

The correlation heatmap for the solvent 2-4PDN shows the relationships between different features in the dataset. The color intensity in the heatmap represents the strength and direction of the correlation. 

Positive values, closer to 1, indicate a positive correlation between the features. This means that as one feature increases, the other feature tends to increase as well. On the other hand, negative values, closer to -1, indicate a negative correlation. This means that as one feature increases, the other feature tends to decrease.

Values close to 0 suggest a weak or no correlation between the features. This means that there is little to no relationship between the changes in one feature and the changes in the other feature.

The colors in the heatmap range from violet, indicating a negative correlation, to blood red, indicating a positive correlation. This color gradient provides a visual representation of the relationships between the features, making it easier to identify patterns and trends in the data.
